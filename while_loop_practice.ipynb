Q1. Print numbers from 1 to 20 using while loop.

Approach:
- Initialize a variable
- Use while condition
- Increment inside loop

```python
#write answer here
var = 1
while var <= 20:
  print(var)
  var += 1
```

Q4. Find sum of numbers from 1 to n.

Approach:
- Take input n
- Add numbers in loop

```python
# Sum of numbers from 1 to n using loop

n = int(input("Enter a number: "))
total = 0

for i in range(1, n+1):
    total += i

print("Sum of numbers from 1 to", n, "is:", total)
```

Q5. Reverse a number using while loop.

Approach:
- Extract last digit
- Build reversed number

```python
n = int(input("enter the number :"))
rev = 0

while n>0:
  digit = n%10
  rev = rev*10+digit
  n= n//10
  print("reversed number is :",rev)
```

Q6. Count digits in a number.

Approach:
- Divide number by 10 repeatedly
- Count iterations

```python
#write answer here
n = int(input("enter the number:"))
count = 0
while n>0:
  n = n//10
  count += 1
  print("the number of digits:",count)
```

Q7. Check if a number is palindrome.

Approach:
- Reverse number
- Compare with original

```python
#write answer here
n = int(input("enter the number:"))
temp = n
rev = 0
while n>0:
  digit = n%10
  rev = rev*10+digit
  n = n//10
if temp == rev:
  print("this is palendrome:")
else:
  print("this is not palendrome:")
```

Q8. Guess the number game.
